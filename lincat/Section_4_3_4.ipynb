# Categories isomorphic to the category of all pairings

This computation is motivated by [arXiv:1904.00166](https://arxiv.org/abs/1904.00166). In particular, it performs the computations of Sect. 4.3.4 of the article.

First let us study the category $\mathscr{K}:=\langle p\rangle_\delta$ generated by the linear combination $p:=\mathsf{abab}+a\,\mathsf{aaaa}$.

```singular
ring r=0,(a,n),dp;
execute(read("lincat.sing"));
addParts(4,gen(7)+a*gen(1));
addTensors();
C[4];
```

From this, we can immediatelly see that the category is non-easy if and only if $a=-2$. As we see in the following computation, the category $\langle\mathsf{abab}-2\,\mathsf{aaaa}\rangle_\delta$ remains to appear non-easy after arbitrary amount of iterations of `addTensors` (actually two iterations are enough since the spaces are already stable afterwards).

```singular
ring r=0,(n),dp;
execute(read("lincat.sing"));
addParts(4,gen(7)-2*gen(1));
```

```singular
addTensors();
printf("Dimensions: (%p,%p,%p,%p)",size(C[2]),size(C[4]),size(C[6]),size(C[8]));
```

```singular
addTensors();
printf("Dimensions: (%p,%p,%p,%p)",size(C[2]),size(C[4]),size(C[6]),size(C[8]));
```

```singular
addTensors();
printf("Dimensions: (%p,%p,%p,%p)",size(C[2]),size(C[4]),size(C[6]),size(C[8]));
```

```singular
C[4];
```

Moreover, writing down the dimensions of the spaces is also helpful to notice that this category might be actually isomorphic to the easy category $\mathbf{Pair}_\delta=\langle\mathsf{abab}\rangle_\delta$ as follows from the following computation.

```singular
ring r=0,(n),dp;
execute(read("lincat.sing"));
addParts(4,gen(7));
```

```singular
addTensors();
printf("Dimensions: (%p,%p,%p,%p)",size(C[2]),size(C[4]),size(C[6]),size(C[8]));
```

```singular
addTensors();
printf("Dimensions: (%p,%p,%p,%p)",size(C[2]),size(C[4]),size(C[6]),size(C[8]));
```

```singular
addTensors();
printf("Dimensions: (%p,%p,%p,%p)",size(C[2]),size(C[4]),size(C[6]),size(C[8]));
```

Note that those numbers are not the true dimensions of the morphism spaces. Due to the length bound, it is very probabl that we do not find all the elements of the category and hence the dimensions will be lower. Indeed, the actual dimensions for the category of all pairings is given by $\dim\mathbf{Pair}_\delta(0,k)=(k-1)!!=1,3,15,105,\dots$ for $k=2,4,6,8,\dots$. Nevertheless, if the given two categories indeed are isomorphic and if take as the generator the corresponding image under this isomorphism, then not only the true dimensions, but also the computed ones must coincide.

The same can be done for another candidate â€“ the category
$$\langle\delta^2\mathsf{abab}-2\delta(\mathsf{abac}+\mathsf{abcb})+4\mathsf{abcd}\rangle_\delta.$$


```singular
ring r=0,(n),dp;
execute(read("lincat.sing"));
addParts(4,n^2*gen(7)-2*n*(gen(8)+gen(13))+4*gen(15));
```
